```python
def fib(n):                               # число фибоначи
    if n <=1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
from math import sqrt                      # квадратные уравнения

def solve(a, b, c):
    d = b**2 - 4*a*c
    if d >= 0:
        x1 = (-b + sqrt(d)) / (2 * a)
        x2 = (-b - sqrt(d)) / (2 * a)   
        return (x1, x2)
    else:
        return 'вещественных корней нет'
```

```python
solve(0.5, 0.125, 0)
```

```python
solve(1, 2, 3)
```

```python
import cmath               # вычисление корней комплексных чисел

def csolve(a, b, c):
    d = b**2 - 4*a*c
    x1 = (-b + cmath.sqrt(d)) / (2 * a)
    x2 = (-b - cmath.sqrt(d)) / (2 * a)   
    return (x1, x2)
```

```python
csolve(1, 2, 5)
```

```python
import matplotlib.pyplot as plt
import random
vx = [1.0, 2.0, 3.0]
vy = [1.0, 2.0, 1.0]         # вершины треугольника
sx = vx [0]
sy = vy [0]                  # начало вершин
px = []
py = []
for i in range(1, 1000):
    r = random.randint(0, 2)
    sx = (sx + vx[r]) / 2
    sy = (sy + vy[r]) / 2
    px.append(sx)
    py.append(sy)
plt.plot(px, py, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = [1.0, 2.0, 3.0]
y = [1.0, 2.0, 1.0]         # вершины треугольника
sx = x [0]
sy = y [0]                  # начало вершин
px = []
py = []
for i in range(1, 100000):
    r = random.randint(0, 2)
    sx = (sx + x[r]) / 2
    sy = (sy + y[r]) / 2
    px.append(sx)
    py.append(sy)
fig, ax = plt.subplots()
# маркеры цветов 'b','g','r','c','m','y','k','w',  RGB: #00ff00
ax.scatter(px, py, c = 'green', s = 1)
ax.set_facecolor('black')
# ax.grid()
# ax.set_xlabel('')
# ax.set_ylabel('')
fig.set_figwidth(15)
fig.set_figheight(10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
count = 10000
x = list(range(1, 7))
y = [0 for i in x]
fig, ax = plt.subplots()
for _ in range(count):
    y[random.randint(0, 5)] += 1
ax.bar(x, y)
y.sort()
# print('{}%'.format(round((y[5] - y[0]) / count *100, 3)))
stat = [f'{i + 1}: {round(y[i] / count *100)}%' for i in range(0, 6)]
print(stat)
plt.show()
```
